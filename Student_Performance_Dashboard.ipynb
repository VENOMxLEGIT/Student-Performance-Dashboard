```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("student_performance_data.csv")
math_marks = np.array(df["Math"])
science_marks = np.array(df["Science"])
english_marks = np.array(df["English"])
computer_marks = np.array(df["Computer"])

# average marks
average_marks = np.array([math_marks.mean(),science_marks.mean(),english_marks.mean(),computer_marks.mean()])

# highest marks
highest_marks = np.array([math_marks.max(),science_marks.max(),english_marks.max(),computer_marks.max()])

# lowest marks
lowest_marks = np.array([math_marks.min(),science_marks.min(),english_marks.min(),computer_marks.min()])

# summary
total_student = len(df)
average_percentage = df["Percentage"].mean()
highest_percentage = df["Percentage"].max()
lowest_percentage = df["Percentage"].min()

# pandas

subject_average = pd.Series(average_marks,
                            index=["Math","Science","English","Computer"])
# grades
grade_distribution = df["Grade"].value_counts().sort_index()

# gender
gender_distribution = df["Gender"].value_counts().sort_index()

# class percentage
class_average = df.groupby("Class")["Percentage"].mean()

# Attendance
attendance = df.groupby("Class")["Percentage"].mean()

# top 10 students
top_students = df.sort_values(by="Percentage",ascending=False).head(10)

# pass/fail
pass_students = len(df[df["Percentage"] >=40])
fail_students = len(df[df["Percentage"] <= 40])

# dashboard
plt.figure(figsize=(18,10))
plt.suptitle("Student Performance Analysis Dashboard", fontsize = 20,fontweight ="bold")


# student summary

plt.subplot(2,3,1)
plt.axis("off")

summary =f"""
total student : {total_student}

average_percentage :{average_percentage:.2f}%

highest_percentage :{highest_percentage:.2f}%

lowest_percentage :{lowest_percentage:.2f}%
"""

plt.text(0.05,0.5,summary,fontsize=15,bbox=dict(facecolor="lightgreen"))

# average_marks
plt.subplot(2,3,2)
plt.bar(subject_average.index,
        subject_average.values)
plt.title("average marks by subject")
plt.ylabel("marks")

# grade distribution
plt.subplot(2,3,3)
plt.pie(grade_distribution.values,labels=grade_distribution.index,
        autopct="1.1f%%",
        startangle=90,
        )
plt.title("Grade distribution")

# class average
plt.subplot(2,3,4)
plt.plot(
    class_average.index,
    class_average.values,
    marker="o",
    linewidth=3
)

plt.title("class-wise average percentage")
plt.ylabel("Percentage")


# gender distribution
plt.subplot(2,3,5)

plt.bar(
    gender_distribution.index,
    gender_distribution.values
)

plt.title("Gender Distribution")
plt.ylabel("Students")

# attendance
plt.subplot(2,3,6)

plt.bar(
    attendance.index,
    attendance.values
)

plt.title("Average Attendance")
plt.ylabel("Attendance (%)")

plt.tight_layout()

plt.show()

# top 10 students

print("\nTOP 10 STUDENTS")
print("-"*70)

print(
    top_students[
        [
            "Student_ID",
            "Name",
            "Class",
            "Percentage",
            "Grade"
        ]
    ]
)

# report
print("\n")
print("="*60)
print("student performance report")
print("="*60)

print("total student :",total_student)
print("average percentage:{:.2f}%".format(average_percentage))
print("highest percentage:{:.2f}%".format(highest_percentage))
print("lowest percentage:{:.2f}%".format(lowest_percentage))

print("\nTOP 10 STUDENTS")
print("-"*70)

print(
    top_students[
        [
            "Student_ID",
            "Name",
            "Class",
            "Percentage",
            "Grade"
        ]
    ]
)






```
